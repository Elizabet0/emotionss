```python
from PIL import Image
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os

p = "../data/train"
classes = list(os.listdir(p))

plt.figure(figsize=(15, 8))
for i, em in enumerate(classes):
    em_p = os.path.join("../data/train", em)
    img_path = os.path.join(em_p, os.listdir(em_p)[0])
    img = Image.open(img_path)
    plt.subplot(2, 4, i+1)
    plt.imshow(img ,  cmap='gray')
    plt.title(f"{em}")
    
plt.show()
```

```python
img_array = np.array(img, dtype=np.float32)
print(f"  : {img_array.shape}")
```

```python
count_train=0
count_test=0
for i, em in enumerate(classes):
    em_p = os.path.join("../data/train", em)
    count_train += len(os.listdir(em_p))

for i, em in enumerate(classes):
    em_p = os.path.join("../data/test", em)
    count_test += len(os.listdir(em_p))
print(f": {count_train} ;  : {count_test}")
```

```python
count_ems_train = {}
for i in os.listdir("../data/train"):
    count_ems_train[i] = len(os.path.join("../data/train", i))

count_ems_test = {}
for i in os.listdir("../data/test"):
    count_ems_test[i] = len(os.path.join("../data/test", i))

plt.figure(figsize=(10, 4))
total_train = sum(count_ems_train.values())
plt.subplot(1, 2, 1)
sns.barplot(count_ems_train)
plt.title('distrib')
for i, count in enumerate(count_ems_train.values()):
    perc = 100 * count / total_train
    plt.text(i, count, f'{perc:.1f}%', ha='center', va='bottom')

plt.subplot(1, 2, 2)
total_test = sum(count_ems_test.values())
sns.barplot(count_ems_test)
plt.title('distrib')
for i, count in enumerate(count_ems_test.values()):
    perc = 100 * count / total_test
    plt.text(i, count, f'{perc:.1f}%', ha='center', va='bottom')

plt.show()
```
